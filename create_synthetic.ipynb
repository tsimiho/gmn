```python
import torch
from custom.utils import *
from custom.dataset import *
import pickle
import os


cycles = torch.load("temp_cycles_new.pt", weights_only=False)
completes, lines, stars = torch.load("temp.pt", weights_only=False)

cycle_prot = cycles[2][0]
line_prot = lines[0][0]
complete_prot = completes[1][0]
star_prot = stars[3][0]
```

```python
gnnexplainer = {"cycle": [], "complete": [], "line": [], "star": []}

for filename in os.listdir("gnnexplainer_graphs"):
    if filename.endswith("cycle.pkl"):
        file_path = os.path.join("gnnexplainer_graphs", filename)
        with open(file_path, "rb") as f:
            original_G, prototype_G = pickle.load(f)
            gnnexplainer["cycle"].append((original_G, prototype_G))

for filename in os.listdir("gnnexplainer_graphs"):
    if filename.endswith("complete.pkl"):
        file_path = os.path.join("gnnexplainer_graphs", filename)
        with open(file_path, "rb") as f:
            original_G, prototype_G = pickle.load(f)
            gnnexplainer["complete"].append((original_G, prototype_G))

for filename in os.listdir("gnnexplainer_graphs"):
    if filename.endswith("line.pkl"):
        file_path = os.path.join("gnnexplainer_graphs", filename)
        with open(file_path, "rb") as f:
            original_G, prototype_G = pickle.load(f)
            gnnexplainer["line"].append((original_G, prototype_G))

for filename in os.listdir("gnnexplainer_graphs"):
    if filename.endswith("star.pkl"):
        file_path = os.path.join("gnnexplainer_graphs", filename)
        with open(file_path, "rb") as f:
            original_G, prototype_G = pickle.load(f)
            gnnexplainer["star"].append((original_G, prototype_G))
```

```python
protgnn = {"cycle": [], "complete": [], "line": [], "star": []}

for filename in os.listdir("protgnn_graphs"):
    if filename.endswith("cycle.pkl"):
        file_path = os.path.join("protgnn_graphs", filename)
        with open(file_path, "rb") as f:
            original_G, prototype_G = pickle.load(f)
            protgnn["cycle"].append((original_G, prototype_G))

for filename in os.listdir("protgnn_graphs"):
    if filename.endswith("complete.pkl"):
        file_path = os.path.join("protgnn_graphs", filename)
        with open(file_path, "rb") as f:
            original_G, prototype_G = pickle.load(f)
            protgnn["complete"].append((original_G, prototype_G))

for filename in os.listdir("protgnn_graphs"):
    if filename.endswith("line.pkl"):
        file_path = os.path.join("protgnn_graphs", filename)
        with open(file_path, "rb") as f:
            original_G, prototype_G = pickle.load(f)
            protgnn["line"].append((original_G, prototype_G))

for filename in os.listdir("protgnn_graphs"):
    if filename.endswith("star.pkl"):
        file_path = os.path.join("protgnn_graphs", filename)
        with open(file_path, "rb") as f:
            original_G, prototype_G = pickle.load(f)
            protgnn["star"].append((original_G, prototype_G))
```

```python
other = (
    []
)  # tuples: (original, protgnn_prototype, gnnexplainer_prototype, ground_truth_prototype)

for i in range(10):
    protgnn_original, protgnn_prot = protgnn["cycle"][i]
    gnnexplainer_original, gnnexplainer_prot = gnnexplainer["cycle"][i]
    other.append(
        (
            protgnn_original,
            protgnn_prot,
            gnnexplainer_prot,
            to_networkx(cycle_prot, to_undirected=True),
        )
    )

for i in range(10):
    protgnn_original, protgnn_prot = protgnn["complete"][i]
    gnnexplainer_original, gnnexplainer_prot = gnnexplainer["complete"][i]
    other.append(
        (
            protgnn_original,
            protgnn_prot,
            gnnexplainer_prot,
            to_networkx(complete_prot, to_undirected=True),
        )
    )

for i in range(10):
    protgnn_original, protgnn_prot = protgnn["line"][i]
    gnnexplainer_original, gnnexplainer_prot = gnnexplainer["line"][i]
    other.append(
        (
            protgnn_original,
            protgnn_prot,
            gnnexplainer_prot,
            to_networkx(line_prot, to_undirected=True),
        )
    )

for i in range(10):
    protgnn_original, protgnn_prot = protgnn["star"][i]
    gnnexplainer_original, gnnexplainer_prot = gnnexplainer["star"][i]
    other.append(
        (
            protgnn_original,
            protgnn_prot,
            gnnexplainer_prot,
            to_networkx(star_prot, to_undirected=True),
        )
    )
```

```python
gmn = []

for i in range(10):
    g, _, _ = cycles[i]
    gmn.append(
        (
            to_networkx(g, to_undirected=True),
            to_networkx(cycle_prot, to_undirected=True),
        )
    )

for i in range(10):
    g, _, _ = completes[i]
    gmn.append(
        (
            to_networkx(g, to_undirected=True),
            to_networkx(complete_prot, to_undirected=True),
        )
    )

for i in range(10):
    g, _, _ = lines[i]
    gmn.append(
        (to_networkx(g, to_undirected=True), to_networkx(line_prot, to_undirected=True))
    )

for i in range(10):
    g, _, _ = stars[i]
    gmn.append(
        (to_networkx(g, to_undirected=True), to_networkx(star_prot, to_undirected=True))
    )
```

```python
with open("synthetic_prototypes.pkl", "wb") as f:
    pickle.dump((gmn, other), f)
```
